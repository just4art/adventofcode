```python
# copy cookie from adventofcode.com to cookie.txt to get inputs
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}
```

# Day 1

```python
url = 'https://adventofcode.com/2025/day/1/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
#1
val = 50 
def parse(x):
    if x[0] == 'R':
        return int(x[1:])
    else: 
        return -int(x[1:])

rot = [parse(x) for x in inp.split('\n')[:-1]]

counter = 0
for i in rot: 
    val = (val + i)%100
    
    if val==0: 
        counter += 1

print(counter)
    
```

```python
#2 
val=50
counter = 0
for i in rot: 
    nval = val + i
    if nval == 0: 
        counter += 1
    if nval < 0 and val!=0: 
        counter += 1
        
    counter += abs(nval)//100
    val = nval%100

print(counter)
```

# Day 2

```python
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}

url = 'https://adventofcode.com/2025/day/2/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
#1
inps = [[int(y) for y in x.split('-')] for x in inp.split(',')]

s = 0
for i in inps: 
    for k in range(i[0],i[1]+1):
        kk = str(k)
        l = len(kk)
        if l%2!=0:
            continue 
        if kk[:l//2]==kk[l//2:]:
            s += k
print(s)
```

```python
#2

s = 0
for i in inps: 
    for k in range(i[0],i[1]+1):
        kk = str(k)
        l = len(kk)
        for j in range(1, l//2+1):
            if l%j==0: 
                c = []
                for h in range(l//j):
                    c.append(kk[j*h:j*(h+1)])
                if len(set(c))==1:
                    s += k
                    break
print(s)
```

# Day 3

```python
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}

url = 'https://adventofcode.com/2025/day/3/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
#1 
inps = [[int(z) for z in x] for x in inp.split('\n')[:-1]]
s = 0
for i in range(len(inps)): 
    max1 = -1 
    ind = -1
    for j in range(len(inps[i])-1): 
        if inps[i][j] > max1: 
            max1 = inps[i][j]
            ind = j 

    max2=0
    for j in range(ind+1, len(inps[i])): 
        if inps[i][j] > max2: 
            max2 = inps[i][j]
    s += 10*max1+max2
print(s)
```

```python
#2 

s = 0
for i in range(len(inps)): 
    num = 0
    ind = -1

    for k in range(11,-1,-1):
        maxx = -1
        for j in range(ind+1, len(inps[i])-k):
            if inps[i][j] > maxx: 
                maxx = inps[i][j]
                ind = j 
        num = num*10 + maxx 
#     print(num, inps[i])
    s += num 
    
print(s)
```
