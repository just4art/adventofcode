# Day 1

```python
url = 'https://adventofcode.com/2025/day/1/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
#1
val = 50 
def parse(x):
    if x[0] == 'R':
        return int(x[1:])
    else: 
        return -int(x[1:])

rot = [parse(x) for x in inp.split('\n')[:-1]]

counter = 0
for i in rot: 
    val = (val + i)%100
    
    if val==0: 
        counter += 1

print(counter)
    
```

```python
#2 
val=50
counter = 0
for i in rot: 
    nval = val + i
    if nval == 0: 
        counter += 1
    if nval < 0 and val!=0: 
        counter += 1
        
    counter += abs(nval)//100
    val = nval%100

print(counter)
```

# Day 2

```python
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}

url = 'https://adventofcode.com/2025/day/2/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
#1
inps = [[int(y) for y in x.split('-')] for x in inp.split(',')]

s = 0
for i in inps: 
    for k in range(i[0],i[1]+1):
        kk = str(k)
        l = len(kk)
        if l%2!=0:
            continue 
        if kk[:l//2]==kk[l//2:]:
            s += k
print(s)
```

```python
#2

s = 0
for i in inps: 
    for k in range(i[0],i[1]+1):
        kk = str(k)
        l = len(kk)
        for j in range(1, l//2+1):
            if l%j==0: 
                c = []
                for h in range(l//j):
                    c.append(kk[j*h:j*(h+1)])
                if len(set(c))==1:
                    s += k
                    break
print(s)
```

# Day 4 

```python
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}

url = 'https://adventofcode.com/2025/day/4/input'
r = requests.post(url, cookies=cookie)
inp = r.text
```

```python
# 1 

inps = inp.split('\n')[:-1]
counter = 0
for i in range(len(inps)):
    for j in range(len(inps[i])):
        if inps[i][j]=='.':
            continue 
        else:
            s = ''.join([x[j-1:j+2] for x in inps[i-1:i+2]])
            s = ''.join([x[max(0, j-1):min(j+2,len(inps[i]))] 
                         for x in inps[max(0,i-1):min(len(inps),i+2)]])

            s = s.replace('.','')
            if len(s)<=4: 
                counter += 1
print(counter)
```

```python
# 2 
counter = 0
res = []


while True: 
    for i in range(len(inps)):
        r = ''
        for j in range(len(inps[i])):
            if inps[i][j]=='.':
                r += '.'
                continue
            else:
                s = ''.join([x[max(0, j-1):min(j+2,len(inps[i]))] for x in inps[max(0,i-1):min(len(inps),i+2)]])
                s = s.replace('.','')
                if len(s)<=4: 
                    r+= '.'
                    counter += 1
                else:
                    r+= '@'
        res.append(r)
    if inps == res: 
        break
    inps = res
    res = []

print(counter)
```

# Day 5

```python
import requests
c = open('cookie.txt').readline()
cookie = {'session': c[:-1]}

url = 'https://adventofcode.com/2025/day/5/input'
r = requests.post(url, cookies=cookie)
inp = r.text
sep = 192
```

```python
# 1 
ranges = inp.split('\n')[:sep]
values = inp.split('\n')[sep+1:]

r = [[int(y) for y in x.split('-')] for x in ranges]
v = [int(x) for x in values[:-1]]
counter = 0
for val in v: 
    for k in r:
        if val >= k[0] and val<= k[1]:
            counter += 1
            break            
print(counter)
```

```python
# 2 
r.sort()
newr = [r[0]]
for s1,s2 in r[1:]: 
    if s1>s2: 
        continue
    for l1,l2 in newr: 
        if s1>=l1 and s1<=l2: 
            s1 = l2 + 1
            if s2 < l2:
                # inside of another range, don't need it
                continue 
        
        elif s2>=l1 and s2<=l2: 
            s2 = l1-1
    if s2>=s1:
        newr.append([s1,s2])
                
counter = 0
for s1,s2 in newr: 
    counter += (s2-s1+1)
    
print(counter)
```
